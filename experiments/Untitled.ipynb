```python
import pandas as pd
%matplotlib inline
```

```python
lookback = pd.read_csv("lookback.csv")
```

```python
lookback
```

```python
plt = lookback.plot(x = "lookback", title="RMSE at Various Lookbacks",y=["rmse_delay","rmse_no_delay"])
plt.set_ylabel("Root Mean Square Error")
plt.set_xlabel("Lookback (min)")
```

```python
sensitivities = pd.read_csv("sensitivities.csv")
```

```python
plt = sensitivities.boxplot(by = "sensitive", column = "rmse_difference")
plt.set_title("Difference Between Sensitive and Insenstive Models\n\n")
plt.set_ylabel("Difference in RMSE \n b/w delay and no delay")
plt.set_xlabel("Model Sensitivity")
```

```python
delays = pd.read_csv("delay.csv")
```

```python
plt = delays.plot(x = "average_delay", y=["rmse_delay","rmse_no_delay"],title = "RMSE at Various Delays")
plt.set_xlabel(xlabel="Average Delay (min)")
plt.set_ylabel(ylabel="RMSE")
```

```python
frequencies = pd.read_csv("frequencies.csv")
```

```python
frequencies["frequency"] = [1,2,5,10,20]
```

```python
plt = frequencies.plot(x = "frequency", y=['rmse_delay','rmse_no_delay'],title = "RMSE at Difference Data Source Frequencies")
plt.set_xlabel("Average Time Between Points (min)")
plt.set_ylabel("RMSE")
```
